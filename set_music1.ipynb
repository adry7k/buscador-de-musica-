# ***Graficos visuais para analises***

# **Horários mais comuns de escuta no dia**

```python
figura_1 = px.histogram(data_music, x='Listening Time (Morning/Afternoon/Night)', title='Horários Mais Comuns de Escuta no Dia')
figura_1.update_layout(bargap=0.2)
figura_1.show()
```

## ***Esse gráfico mostra em qual parte do dia em que as pessoas mais consomem música. Sendo a “Noite” foi a mais alta,  demonstra um padrão de escuta noturna. Isso ajuda plataformas a entender quando promover conteúdo ou playlists para os usuários.***

## **Top 5 artistas mais ouvidos por país**

```python

top_paises = data_music['Country'].value_counts().head(5).index
data_top_artistas = data_music[data_music['Country'].isin(top_paises)]
artistas_por_pais = data_top_artistas.groupby(['Country', 'Most Played Artist']).size().reset_index(name='Contagem')
figura_2 = px.bar(artistas_por_pais, x='Most Played Artist', y='Contagem', color='Country', barmode='group', title='Artistas Mais Ouvidos por País (Top 5)')
figura_2.show()
```

# **Esse gráfico permite observar preferências culturais e regionais. Por exemplo, um artista pode ser muito popular no Brasil, mas quase irrelevante em outros lugares.**

# **Qual é a média do gênero musical mais ouvido por faixa etária**

```python

genero_faixa = data_music.groupby(['Faixa_Etaria', 'Top Genre']).size().reset_index(name='Contagem')
max_genero_faixa = genero_faixa.sort_values('Contagem', ascending=False).drop_duplicates('Faixa_Etaria')
figura_3 = px.bar(max_genero_faixa, x='Faixa_Etaria', y='Contagem', color='Top Genre', title='Média do Gênero Musical Mais Ouvido por Faixa Etária')
figura_3.show()
```

Ao invés de apenas apontar o mais ouvido, calculamos a média de ouvintes por faixa etária para cada gênero predominante.

## ***Aqui entendemos quais gêneros são preferidos por faixa etária, por exemplo: jovens adultos preferem EDM (música eletrônica) e os adultos gostam de reggae.***

# **Quantidade de Assinantes Premium por Plataforma**

```python

premium_por_plataforma = data_music.groupby(['Streaming Platform', 'Subscription Type']).size().reset_index(name='Contagem')
premium_por_plataforma['Subscription Type'] = premium_por_plataforma['Subscription Type'].astype(bool)
premium_true = premium_por_plataforma[premium_por_plataforma['Subscription Type'] == True]
figura_4 = px.pie(premium_true, names='Streaming Platform', values='Contagem',
             color='Streaming Platform', title='Quantidade de Assinantes Premium por Plataforma')
figura_4.show()

```

```python
agrupado_faixa = data_music.groupby('Faixa_Etaria')[['Repeat Song Rate (%)', 'Discover Weekly Engagement (%)']].mean().reset_index()
figura_5 = px.bar(agrupado_faixa, x=['Repeat Song Rate (%)', 'Discover Weekly Engagement (%)'], y='Faixa_Etaria', barmode='group', title='Fidelidade e Descoberta por Faixa Etária', orientation="h")
figura_5.show()
```
